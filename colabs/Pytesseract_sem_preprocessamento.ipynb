# Passo 01: Listar todos diretórios

```python
path_diretorio_txt = '/content/drive/MyDrive/03_visagio/desafio_python/txt' #path_diretorio_txt o caminho para a pasta que receberá o arquivo com os txts das imagens
path_diretorio_confiabilidade = '/content/drive/MyDrive/03_visagio/desafio_python/confiabilidade' #path_diretorio_confiabilidade o caminho para a pasta que receberá o arquivo com as confiabilidades
path_diretorio_imagens = '/content/drive/MyDrive/03_visagio/desafio_python/imagens2' #path_diretorio_imagens o caminho para a pasta que contem todas as imagens
lista_arquivos = os.listdir(path_diretorio_imagens)
```

## LOOP: contem passos 2,3,4,5,6,7



*   Passo 2: Lê a imagem
*   Passo 3: Analisa o tamanho da imagem
*   Passo 4: Usar o pytesseract para ler o texto da imagem
*   Passo 5: Confiablidade
*   Passo 6: Salvar em um arquivo com o nome do arquivo da imagem:
*   Passo 7: Salva a confiabilidade da palavra
*   Passo 8: move o arquivo ara pasta de destino

```python
for arquivo in lista_arquivos:

  # Passo 2: Ler a imagem
  path_imagem = path_diretorio_imagens + '/' + arquivo
  imagem = cv2.imread(path_imagem)

  # Passo 3: Analisa tamanho da imagem
  tamanho = imagem.shape
  if (tamanho[0] <= 800 and tamanho[1] <= 800):
    
    scale = 800/(min(tamanho[0],tamanho[1]))
    new_height = int(tamanho[0]*scale)
    new_width = int(tamanho[1]*scale)

    #redimensionar
    imagem = cv2.resize(imagem, (new_width, new_height))

  # Passo 4: Usar o pytesseract para ler o texto da imagem
  config_tesseract = '--tessdata-dir tessdata'
  texto = pytesseract.image_to_string(imagem, lang='por', config=config_tesseract)

  # Passo 5: confiabilidade
  informacoes = pytesseract.image_to_data(imagem, output_type='data.frame')

  #calculo da confiabilidade
  soma = 0
  count = 0
  for conf in informacoes.conf:
    if conf > 0:
      soma += conf
      count += 1

  # Passo 6: salvar em um arquivo com o nome do arquivo da imagem:
  nome_arquivo_imagem = path_imagem.split('/')[-1]
  nome_arquivo = nome_arquivo_imagem.split('.')[0] + ".txt"
  arquivo_imagem = open(nome_arquivo,'w')
  arquivo_imagem.write(texto)
  arquivo_imagem.close()

  # Passo 7: Salva a confiabilidade da palavra
  arquivo_conf = open('confiabilidade.txt','a')
  confiabilidade = f"{nome_arquivo_imagem} {soma/count}\n"
  print(f"Arquivo: {nome_arquivo_imagem} lido! Confiabilidade:{soma/count}")
  arquivo_conf.writelines(confiabilidade)
  arquivo_conf.close()

  # Passo 8: move o arquivo ara pasta de destino
  shutil.move(nome_arquivo, path_diretorio_txt)

shutil.move('confiabilidade.txt', path_diretorio_confiabilidade)
```
