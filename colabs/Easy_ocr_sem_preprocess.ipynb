```python
# importando bibliotecas necessárias

import os
from easyocr import Reader
import cv2
from google.colab.patches import cv2_imshow
from PIL import ImageFont, ImageDraw, Image
import numpy as np
import shutil
```

```python
#listar idiomas

idiomas = ['en','pt']
```

```python
#Ativar o uso da GPU

gpu = True
```

# Passo 01: Listar todos os arquivos no diretório que contem as imagens

```python
#coloque na variável path_diretorio o caminho para a pasta que contem todas as imagens

path_diretorio = '/content/imagens_ocr'
lista_arquivos = os.listdir(path_diretorio)
```

# Passo 02: Inserir caminho do diretório onde vai ser guardado os arquivos txt gerados na leitura.

```python
# Coloque na variável path_destino o caminho para a pasta que vai armazenar as transcrições em txt.
# Caso não tenha criado uma pasta, o próprio programa irá criar uma pasta chamada "resultados".

path_destino = ''

# Aqui vai criar a pasta resultado, caso não seja inserido nenhum caminho.
if not path_destino:
  dir = '/content/resultados' 
  os.makedirs(dir)
  path_destino = '/content/resultados'
```

## LOOP: contem passos 2,3,4,5,6,7



*   Passo 2: Lê a imagem
*   Passo 3: Analisa o tamanho da imagem
*   Passo 4: Usa o easyocr para ler o texto da imagem
*   Passo 5: Printa e concatena o texto
*   Passo 6: Salvar em um arquivo com o nome do arquivo da imagem:
*   Passo 7: Salva a confiabilidade da palavra
*   Passo 8: move o arquivo ara pasta de destino


```python
for arquivo in lista_arquivos:

  # Passo 2: Ler a imagem
  path_imagem = path_diretorio + '/' + arquivo
  imagem = cv2.imread(path_imagem)
  
  # Passo 3: Analisa tamanho da imagem
  tamanho = imagem.shape
  if (tamanho[0] <= 800 and tamanho[1] <= 800):
    
    scale = 800/(min(tamanho[0],tamanho[1]))
    new_height = int(tamanho[0]*scale)
    new_width = int(tamanho[1]*scale)

    #redimensionar
    imagem = cv2.resize(imagem, (new_width, new_height))

  # Passo 4: Usar o easyocr para ler o texto da imagem
  reader = Reader(idiomas, gpu)
  resultados = reader.readtext(imagem)

  # Passo 5: Printar o texto
  soma = 0
  count = 0
  texto =''
  for i in range(len(resultados)):
    soma = soma + resultados[i][2]
    count += 1
    texto = texto + str(resultados[i][1]) + ' '

  # Passo 6: salvar em um arquivo com o nome do arquivo da imagem:
  nome_arquivo_imagem = path_imagem.split('/')[-1]
  nome_arquivo = nome_arquivo_imagem.split('.')[0] + ".txt"
  arquivo_imagem = open(nome_arquivo,'w')
  arquivo_imagem.write(texto)
  arquivo_imagem.close()

  # Passo 7: Salva a confiabilidade da palavra
  arquivo_conf = open('confiabilidade.txt','a')
  confiabilidade = f"{nome_arquivo_imagem} {soma/count:.2%}\n"
  print(f"Arquivo: {nome_arquivo_imagem} lido! Confiabilidade:{soma/count:.2%}")
  arquivo_conf.writelines(confiabilidade)
  arquivo_conf.close()

  # Passo 8: move o arquivo ara pasta de destino
  shutil.move(nome_arquivo, path_destino)
```
